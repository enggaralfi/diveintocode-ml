The problem should be the manju cake is throwing to the universe, then the volume of the cake will expand
2 times larger.
Let assume that the volume of manju before expanding is x and the volume of universe is y. manju will expand 2 time larger after 5 minutes. 
The problem is how long manju will covers to all universe? let assume i as how many time expanding.
Mathematical formula should be: Y=x\*2^i. if we get i, then just times by 5 minutes.

```python
import numpy as np
import matplotlib.pyplot as plt
def volume_expanding(v_0=3e-6,v_final=1.4e27,t_expand=5.):
    vol=0
    i=0
    t_run=[]
    vol_run=[]
    while vol < v_final:
        i+=1
        vol=(v_0*(2**i))
        t=(i*t_expand)
        t_run.append((i*t_expand))
        vol_run.append((v_0*(2**i)))
    %matplotlib inline
    plt.title("Expanding volume")
    plt.xlabel("time(minutes)")
    plt.ylabel("Volume (m^3)")
    plt.tick_params(labelsize=12)
    plt.plot(t_run,vol_run, color='green', marker='^', linestyle='dashed')
    plt.show()
    return vol,t
    
vmanju,tmanju=volume_expanding()    
print("The manju will close the sun after reach volume {} m^3 in time {} minutes".format(vmanju,tmanju))
```

```python
#Ball covers the tokyo dome
"""
In this case we will calculate of expanding ball (5546.44 cubic centimeters) will covers 
the Tokyo Dome (1.24e6)cubic meter.
"""

vTDome,tTDome=volume_expanding(v_0=5546.44e-6,v_final=1.24e6,t_expand=5)
print("the ball will cover the Tokyo Dome when reach the volume {} m^3 in time {} minutes".format(vTDome,tTDome))
```
