Create 500 random numbers with a two-dimensional normal distribution whose mean is (-3, 0) and whose covariance matrix is


\[\left[ \begin{array}{ccc} 1.0 & 0.8\\ 0.8 & 1.0 \end{array} \right]\]

```python
import numpy as np
mean = (-3,0)
cov = [[1.,.8],[.8,1.]]
x1,x2=np.random.multivariate_normal(mean,cov,(500,2)).T
print(x1)
print(x2)
```

# Problem 2. Plot the data

```python
import matplotlib.pyplot as plt
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Scatter")
plt.scatter(x1,x2)
```

# Problem 3. Visualisation by histogram

```python
plt.xlabel("X1")
plt.ylabel("Frequancy")
plt.title("Histogram of X1")
plt.hist(x1)
```

```python
plt.xlabel("X2")
plt.ylabel("Frequancy")
plt.title("Histogram of X2")
plt.hist(x2)
```

# Problem 4. additional of data

```python
mean_1 = (0,-3)
cov_1 = [[1.,.8],[.8,1.]]
x3,x4=np.random.multivariate_normal(mean_1,cov_1,(500,2)).T
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Scatter")
plt.scatter(x1,x2)
plt.scatter(x3,x4)
plt.show()
```

# Problem 5. Data Combining

```python
x1_combine=np.concatenate((x1,x3))
x2_combine=np.concatenate((x2,x4))
x1_reshape_1000_2 = x1_combine.reshape(1000,2)
x2_reshape_1000_2 = x2_combine.reshape(1000,2)
print(x1_combine.shape)
```

# Problem 6. labelling

```python
zero_label = np.zeros(500)
one_label = np.ones(500)
label_all = np.concatenate((zero_label,one_label))
label_1000_2 = label_all.reshape(1000,1)
x1_labeled = np.concatenate((x1_reshape_1000_2,label_1000_2),axis=1) 
print(x1_labeled)
```
